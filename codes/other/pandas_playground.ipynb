### **Pandas Series**

```python
# Get a column as Series
temps = df['temp']

# Get a row as series
r0 = df.iloc[0]

print('\033[35;1m temps column: \033[0m')
print(temps)
print('-'*20)
print('\033[35;1m first row: \033[0m')
print(r0)
```

```python
print('Mean:', temps.mean())
print('STD:', temps.std())
```

### Conditional Indexing

```python
# SELECT * FROM df WHERE condition='Rain';
df2 = df[df['condition'] == 'Rain']
df2
```

```python
# SELECT * FROM df WHERE temp=(SELECT MAX(temp) FROM df);
df3 = df[df['temp'] == df['temp'].max()]
df3
```
